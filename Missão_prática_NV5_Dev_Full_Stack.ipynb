```python
from faker import Faker
import random
import matplotlib.pyplot as plt
from num2words import num2words
from wordcloud import WordCloud
```

```python
#geração de dados
f = Faker('pt-BR')
numeros=[]
nomes=[]
with open('alunos.txt', 'w') as arquivo:


  for i in range(10):
    notas=f.name()+', '+str(f.random_int(1,10))+' \n'
    arquivo.write(notas)
    print(notas)
```

```python
#recuperar dados
with open('alunos.txt', 'r') as arquivo:
  lista_separar=[]
  for line in range(10):
    lista=arquivo.readline()
    separar=lista.split(' ')
    lista_separar.append(int(separar[-2]))
  print(lista_separar)
```

```python
plt.title('Histograma das pontuações')
plt.xlabel('Pontuações')
plt.ylabel('Probabilidade')
num_bins=10; plt.hist(lista_separar,num_bins, density=True,facecolor='red', rwidth=0.8)
plt.grid(True)
plt.show()
```

```python
nova_lista=[]
for i in lista_separar:
  notas_escrita=num2words(i, lang='pt_br')
  nova_lista.append(notas_escrita)  
cloud_notas=(',').join(nova_lista)
print(cloud_notas)
```

```python
nuvem=WordCloud().generate(cloud_notas)
plt.imshow(nuvem, interpolation='bilinear')
plt.axis('off')
plt.show()
nuvem.to_file('nuvem de palavras.png')
```
